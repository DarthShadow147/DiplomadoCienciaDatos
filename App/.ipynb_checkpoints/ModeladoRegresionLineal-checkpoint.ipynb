# REGRESIÓN LINEAL SIMPLE

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
train_data,test_data=train_test_split(df,train_size=0.8,random_state=3)
reg=linear_model.LinearRegression()
x_train=np.array(train_data['sqft_living']).reshape(-1,1)
y_train=np.array(train_data['price']).reshape(-1,1)
reg.fit(x_train,y_train)

x_test=np.array(test_data['sqft_living']).reshape(-1,1)
y_test=np.array(test_data['price']).reshape(-1,1)
pred=reg.predict(x_test)
print('Modelo lineal')
mean_squared_error=metrics.mean_squared_error(y_test,pred)
print('Error medio cuadrático', round(np.sqrt(mean_squared_error),2))
print('Entrenamiento R cuadrado',round(reg.score(x_train,y_train),3))
print('Prueba de R cuadrado',round(reg.score(x_test,y_test),3) )
print('Interceptar',reg.intercept_)
print('Coeficiente',reg.coef_)
```

```python
_, ax = plt.subplots(figsize= (12, 10))
plt.scatter(x_test, y_test, color= 'darkgreen', label = 'Datos')
plt.plot(x_test, reg.predict(x_test), color='red', label= ' Línea de regresión pronosticada')
plt.xlabel('Espacio habitable (pies cuadrados)')
plt.ylabel('Precio')
plt.legend()
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```

```python
train_data,test_data=train_test_split(df,train_size=0.8,random_state=3)
reg=linear_model.LinearRegression()
x_train=np.array(train_data['grade']).reshape(-1,1)
y_train=np.array(train_data['price']).reshape(-1,1)
reg.fit(x_train,y_train)

x_test=np.array(test_data['grade']).reshape(-1,1)
y_test=np.array(test_data['price']).reshape(-1,1)
pred=reg.predict(x_test)
print('Modelo lineal')
mean_squared_error=metrics.mean_squared_error(y_test,pred)
print('Error medio cuadrático',round(np.sqrt(mean_squared_error),2))
print('Entrenamiento R cuadrado',round(reg.score(x_train,y_train),3))
print('Prueba de R cuadrado',round(reg.score(x_test,y_test),3))
print('Interceptar',reg.intercept_)
print('Coeficiente',reg.coef_)
```

# Regresión múltiple

```python
fig,ax=plt.subplots(2,1,figsize=(15,10))
sns.boxplot(x=train_data['grade'],y=train_data['price'],ax=ax[0])
sns.boxplot(x=train_data['bedrooms'],y=train_data['price'],ax=ax[1])
_ , axes = plt.subplots(1, 1, figsize=(15,10))
sns.boxplot(x=train_data['bathrooms'],y=train_data['price'])
```

```python
features1=['bedrooms','grade','sqft_living','sqft_above']
reg=linear_model.LinearRegression()
reg.fit(train_data[features1],train_data['price'])
pred=reg.predict(test_data[features1])
print('Modelo Complejo 1')
mean_squared_error=metrics.mean_squared_error(y_test,pred)
print('Error Medio Cuadrado(MSE)', round(np.sqrt(mean_squared_error),2))
print('Entrenamiento R cuadrado',round(reg.score(train_data[features1],train_data['price']),3))
print('Entrenamiento R cuadrado', round(reg.score(test_data[features1],test_data['price']),3))
print('Interceptar: ', reg.intercept_)
print('Coeficiente:', reg.coef_)
```

```python
features1 = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','grade','sqft_above','sqft_basement','lat','sqft_living15']
reg= linear_model.LinearRegression()
reg.fit(train_data[features1],train_data['price'])
pred = reg.predict(test_data[features1])
print('Modelo Complejo 2')
mean_squared_error = metrics.mean_squared_error(y_test, pred)
print('Error Medio Cuadrado(MSE) ', round(np.sqrt(mean_squared_error), 2))
print('Entrenamiento R cuadrado ', round(reg.score(train_data[features1], train_data['price']), 3))
print('Prueba R cuadrado', round(reg.score(test_data[features1], test_data['price']), 3))
print('Interceptar: ', reg.intercept_)
print('Coeficiente: ', reg.coef_)
```

# Regresión polinomial

```python
polyfeat=PolynomialFeatures(degree=2)
xtrain_poly=polyfeat.fit_transform(train_data[features1])
xtest_poly=polyfeat.fit_transform(test_data[features1])

poly=linear_model.LinearRegression()
poly.fit(xtrain_poly,train_data['price'])
polypred=poly.predict(xtest_poly)

print('Modelo Complejo 3')
mean_squared_error = metrics.mean_squared_error(test_data['price'], polypred)
print('Error Medio Cuadrado(MSE) ', round(np.sqrt(mean_squared_error), 2))
print('Entrenamiento R cuadrado', round(poly.score(xtrain_poly, train_data['price']), 3))
print('Prueba R cuadrado', round(poly.score(xtest_poly, test_data['price']), 3))
```

```python
polyfeat=PolynomialFeatures(degree=3)
xtrain_poly=polyfeat.fit_transform(train_data[features1])
xtest_poly=polyfeat.fit_transform(test_data[features1])

poly=linear_model.LinearRegression()
poly.fit(xtrain_poly,train_data['price'])
polypred=poly.predict(xtest_poly)

print('Modelo Complejo 4')
mean_squared_error=metrics.mean_squared_error(test_data['price'],polypred)
print('Error Medio Cuadrado (MSE) ', round(np.sqrt(mean_squared_error), 2))
print('Entrenamiento R cuadrado', round(poly.score(xtrain_poly, train_data['price']), 3))
print('Prueba R cuadrado ', round(poly.score(xtest_poly, test_data['price']), 3))
```

# Observación/Resultado

El modelo complejo 3 nos da una puntuación de R-cuadrado (prueba) de 0,759. De los informes anteriores, podemos concluir que la regresión polinomial es la mejor solución.
